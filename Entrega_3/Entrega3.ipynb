#**HALLAR EL PUNTO MÁS CERCANO**

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
lats = nc.variables['latitude'][:]
lons = nc.variables['longitude'][:]
time = nc.variables['time'][:]
print(np.shape(lats))
print(np.shape(lons))
print(np.shape(time))
NC_Estaciones = {'Latitud':lats,'Longitud':lons,'Tiempo':time}
# NC_Estaciones = {'Latitud':lats,'Longitud':lons,'Valores':values,'Tiempo':time}
NC_Estaciones
```

```python
# values = NC_Estaciones['Valores']
```

```python
lats_stations = []
lons_stations = []
for station in df_stations['lat']:
  Elat = station
  lats_stations.append(find_nearest(lats,Elat))
for station in df_stations['lon']:
  Elon = station
  lons_stations.append(find_nearest(lons,Elon))
np.array(lats_stations)
np.array(lons_stations)
#-
print(lats_stations)
print(lons_stations)
```

```python
Estaciones = {'Lats':lats_stations,'Lons':lons_stations,'Estacion':df_stations['Estacion'],'Municipio':df_stations['Municipio']}
Estaciones
```
